```python
# Import recommended libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read and show file
df= pd.read_csv('/content/drive/MyDrive/datasets/housing_price_dataset.csv')
df.head()
```

```python
#Check data information (datatype,nullvalues)
df.info()
```

```python
print(df.describe())

print(df.columns)
```

```python

#Heatmap for null values
df1=df.drop("Neighborhood", axis='columns')
sns.heatmap(df1.isna())
```

```python
# Calculate price per square foot
df['price_per_squarefoot'] = df['Price'] / df['SquareFeet']
print("Mean of price_per_squarefoot",df.groupby('Neighborhood')['price_per_squarefoot'].mean())

# Display the DataFrame
print(df.groupby('Neighborhood').count())
```

```python
# create bar plot by  Price per Squarefeet with Neighborhood

df.groupby(['Neighborhood'])['price_per_squarefoot'].mean().plot(kind='bar')
plt.xlabel('Neighborhood')
plt.ylabel('price_per_squarefoot')
plt.title('Price per Squarefeet by Neighborhood')
plt.show()
```

```python
# Assuming your DataFrame is named df
neighborhood_counts = df['Neighborhood'].value_counts()
total_properties = len(df)

# Calculate percentage for each neighborhood
neighborhood_percentage = (neighborhood_counts / total_properties) * 100

# Display the result
print(neighborhood_percentage)
```

```python
# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(neighborhood_percentage, labels=neighborhood_percentage.index, autopct='%1.1f%%', startangle=90)
plt.title('Neighborhood Distribution')
plt.show()
```
